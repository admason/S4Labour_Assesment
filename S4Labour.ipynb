```python
# import libraries
import numpy as np
import pandas as pd
#import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
# Import csv to python

# March 2019
df2019 = pd.read_csv('March_2019_CSV.csv',index_col=0) 
# March 2021
df2021 = pd.read_csv('March_2021_CSV.csv',index_col=0)
#df2019.head()
#df2021
#df1.info()
```

```python
# Combine dataframes for 2019 and 2021 into dataframe labelled  dfAll

dfAll=pd.concat([df2019,df2021])
dfAll.head()
```

```python
round(dfAll['ActualSalesValue'].mean(),0)
```

```python
round(dfAll['ActualSalesValue'].max(),0)
```

```python
sum(dfAll[dfAll['OrganisationID']==281]['ActualSalesValue'])
```

```python
# Top 5 Sales
sales_desc=dfAll.sort_values('ActualSalesValue',axis=0, ascending=False).head(5)
sales_desc
```

# Analysis by OrganisationID

```python
print('Organisation responsible for highest sale: ')
toporg=dfAll[dfAll['ActualSalesValue']==dfAll['ActualSalesValue'].max()]
toporg
```

```python
print('Organisation responsible for lowest sale: ')
bottomorg=dfAll[dfAll['ActualSalesValue']==dfAll['ActualSalesValue'].min()]
bottomorg
```

```python
print('Maximum SalesValue for each OrganisationID')
m=dfAll.groupby('OrganisationID')['ActualSalesValue','tbl_ActualSales.PhysicalSiteID','Type'].max()
m.head()
```

```python
round(m.describe(),0)
```
